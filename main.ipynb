```python
# Thư viện
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
#TASK 1: Tải file .csv lên trương trình
path = 'data/'
df = pd.read_csv(path + "sales2019_1.csv") 
df.head(11)
```

```python
# TASK 2: 
# TASK 2.1: Tải dữ liệu 12 file .csv lên trương trình -> Gộp thành 1 file lớn

all_length = []
frames = [] 
# Danh sách tên file trong thư mục
for file in os.listdir(path):
    if file.endswith(".csv"):
            df = pd.read_csv(path + file)
            frames.append(df)
            result = pd.concat(frames)
            all_length.append(len(df.index))
            

df = result
# Ghi ra 1 file mới dữ liệu của 12 tháng
df.to_csv('annualSale2019.csv', index=False)
```

```python
# So sánh số lượng dòng trước và sau khi gộp
sum(all_length)
```

```python
# TASK 2.2: Thêm cột tháng vào dữ liệu
df['Month'] = df['Order Date'].str[0:2]
df.head(11)
```

```python
# Task 2.3: Làm sạch dữ liệu 
df = df.dropna(how='all')
df = df[df['Month'] != 'Or']
df.head(11 )
```

```python
# Kiểm tra các giá trị trong cột month
len(set(df['Month']))
```

```python
# Task 3: Tháng nào có doanh số tốt nhất, doanh số tháng đó là bao nhiêu?
# Chuyển đổi str to float để tính phép tính
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast='integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast='float') 

# Tạo cột mới "Số tiền thu được của từng đơn hàng" = Quantity Ordered * Price Each
df['Sales'] =  df['Quantity Ordered'] * df['Price Each']
df.head(11)


  
```

```python
# Lọc dữ liệu theo từng tháng
sales_value = df.groupby('Month').sum()["Sales"]
sales_value.max()

```

```python
# Vẽ biểu đồ sales
months = range(1,13)
plt.bar(x = months, height=sales_value)
plt.xticks(months)
plt.xlabel("Months")
plt.ylabel('Sales')
plt.show()
```

```python
# Task 3.2: Thành phố nào có doanh số cao nhất

# Viết hàm xử lý lấy tên thành phố trong địa chỉ
addresses_to_city =  lambda addresses:addresses.split(',')[1]
df['City'] = df['Purchase Address'].apply(addresses_to_city )
```

```python
df.groupby('City').sum()['Sales']
```

```python
sales_value_city = df.groupby('City').sum()['Sales']
sales_value_city.max()
```

```python
cities = [city for city, sales in sales_value_city.items()]
cities

```

```python
# Vẽ biểu đồ sales

plt.bar(x = cities, height=sales_value_city)
plt.xticks(cities, rotation = 45)
plt.xlabel("Cities")
plt.ylabel('Sales')
plt.show()
```

```python
# Task 3: Trả lời câu hỏi: Nên quảng cáo ở khung thời gian nào? Để tăng doanh thu

# Chuyển đổi cột Order Date	dạng string sang datime
df['Order Date'] = pd.to_datetime(df['Order Date'])
print(df['Order Date'].dtypes)
```

```python
# Lấy khung giờ trong Order Date
df['Hours'] = df['Order Date'].dt.hour 
df.head(11)
```

```python
# Vẽ biểu đồ
sales_value_hours = df.groupby('Hours').count()["Sales"]
hours = [hour for hour, sales in sales_value_hours.items()]
plt.grid()
plt.plot( hours, sales_value_hours)
plt.xticks(hours)
plt.xlabel("Hours")
plt.ylabel('Sales')
plt.show()
```
